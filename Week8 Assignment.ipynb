```python
df = df.dropna(subset=["date"])
df["date"] = pd.to_datetime(df["date"])
df["date"] = df["aged_65_older"].fillna(df["aged_70_older"].mean()) 
print(df.head())
print(df.info())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# line chart
plt.figure(figsize=(10, 5))
sns.lineplot(x="new_cases", y="new_deaths", data=df)
plt.title("cases & deaths over time")
```

```python
# Bar chart
plt.figure(figsize=(10, 5))
sns.barplot(x="continent", y="new_cases", data=df)
plt.title("top Continent by total cases")
```
